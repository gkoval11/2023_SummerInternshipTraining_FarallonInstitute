## What is GitHub?

<img src='./figures/github_logo.png' width=200>

At a high level, __GitHub__ is a website and cloud-based service that helps developers store and manage their code, as well as track and control changes to their code. 

__GitHub__ is used for version control to help developers track and manage changes to a software projectâ€™s code. Version control lets developers safely work through branching and merging.

* With branching, a developer duplicates part of the source code (called the repository). The developer can then safely make changes to that part of the code without affecting the rest of the project.

* Then, once the developer gets his or her part of the code working properly, he or she can merge that code back into the main source code to make it official.

All of these changes are then tracked and can be reverted if need be. For our use-case, we'll probably be using GitHub as a way to share and document code.

## The Basics

__GitHub__ uses repositories (basically a project folder) where you can store files such as scripts, data, or figures. Repositories can either be public (anyone can see the files within the repository) or private (only you or invited collaborators can access it). Good practice is to include a `README` markdown file within each repository that details the goals of the repositiory, the file layout, how to get started using the repository, and contact information if someone wants to reach out to you. 

One of the reasons __GitHub__ is so popular with developers is the collaborative potential. This can be done through branches, or a workspace where you can make changes without effecting the whole project. Each repository has a main/master branch where all the main changes are stored. You can use branches to create and store changes which can then be amended to the main branch that then effects the final project. 

When working with __GitHub__, you have to commit files from your local machine and push them to the repository. Each commit is tracked so if you make a mistake or need to revert to a previous version of your project, you can easily do so. Before files are committed, they need to be added to the staging area. The staging area allows you to view the changes that you're about to make to ensure everything looks correct and you didn't forget any files. Below is the typical workflow of __GitHub__. 

On public repositories, people can add comments to the "issues" section where they find bugs or problems while running the code and then developers can address those comments in updates.

### Basic GitHub commands

* `git init` initializes a brand new Git repository and begins tracking an existing directory. It adds a hidden subfolder within the existing directory that houses the internal data structure required for version control.

* `git clone` creates a local copy of a project that already exists remotely. The clone includes all the project's files, history, and branches.

* `git add` stages a change. Git tracks changes to a developer's codebase, but it's necessary to stage and take a snapshot of the changes to include them in the project's history. This command performs staging, the first part of that two-step process. Any changes that are staged will become a part of the next snapshot and a part of the project's history. Staging and committing separately gives developers complete control over the history of their project without changing how they code and work.

* `git commit` saves the snapshot to the project history and completes the change-tracking process. In short, a commit functions like taking a photo. Anything that's been staged with git add will become a part of the snapshot with git commit.

* `git status` shows the status of changes as untracked, modified, or staged.

* `git branch` shows the branches being worked on locally.

* `git merge` merges lines of development together. This command is typically used to combine changes made on two distinct branches. For example, a developer would merge when they want to combine changes from a feature branch into the main branch for deployment.

* `git pull` updates the local line of development with updates from its remote counterpart. Developers use this command if a teammate has made commits to a branch on a remote, and they would like to reflect those changes in their local environment.

* `git push` updates the remote repository with any commits made locally to a branch.

## Ways to Use GitHub

### Brower UI

More user friendly than the command line, but doesn't allow the same control of the repositories. You can still upload or create files, but this interface is more visual to look at the README, download the repository, or other actions related to the repository. It's harder to work with more files and larger changes to the files. 

<img src='./figures/GitHubUI.png' width = 1000>

### Command Line

Accessed through the terminal (Mac/Linux) or through Git (Windows/Mac/Linux). Using the command line is the most abstract, less user-friendly approach to __GitHub__, but allows the most control over everything. Typical command line workflow includes navigating to the directory where your project folder is or where you want it to be:

`cd ~/Documents #cd (change directory) moves to the selected folder on your machine`

Cloning (downloading) an existing repository:

`git clone [URL] #replace the [URL] with the url of the repository`

Initializing that file as a GitHub repository and then making the changes necessary until you are ready to upload: 

`git init . #initializes the folder so you can upload files from there`

Create a remote repository so you can upload data locally:

`git remote add origin [URL] #replace the [URL] with the url of the repository you want to add to and it will be called "origin"`

If you already have a remote repository, you can change the url of the repository with: 

`git remote set-url origin [URL] #replace the [URL] with the url of the new repository you want to add to`

Pull files from the repository so there aren't any issues when you try to commit:

`git pull remote #downloads and merges changes with your files`

Add the files you have to upload to the staging area:

`git add . #adds all files in the current directory`

`git add [FILENAME or FOLDER] #if you only want to add one file or a specific folder`

Commit those files with a message:

`git commit -m "[MESSAGE]" #replace the [MESSAGE] with the description of the changes you made`

And finally push those files to the repository/branch:

`git push origin main #to push to the main branch, you can replace main with another branch you want to push to`

### GitHub Desktop

A hybrid of the command line and browser. Still provides the functionality of the command line but more user friendly. Interface allows you to download repositories, work within those while it tracks the changes, and then commit and push the code from there. Probably the most convenient way to interact with __GitHub__.
